# Preprocessing using nltk

```python
import nltk
import matplotlib.pyplot as plt
import random
```

```python
with open('haku_journal.txt', 'r') as file:
    text_data = file.read()
```

```python
text_data
```

## apply autocorrect on english words

```python
from langdetect import detect
from spellchecker import SpellChecker
```

```python
spell = SpellChecker()

corrected_sentences = []
for sentence in text_data.split('。'):  # Split by sentence (adjust for other languages)
    if detect(sentence) == 'en':  # Detect English sentences
        corrected_sentence = " ".join(
            [spell.correction(word) if spell.correction(word) is not None else word for word in sentence.split()]
        )
        corrected_sentences.append(corrected_sentence)
    else:
        corrected_sentences.append(sentence)  # Leave non-English sentences unchanged

# Combine corrected sentences
corrected_text = '。'.join(corrected_sentences)
print("Corrected Text:", corrected_text)
```

## Tokenizing text data

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
# Run this line only once
# nltk.download('punkt')
tokens = word_tokenize(corrected_text)
print("Tokens:", tokens)
```

## Removing Stopwords and Punctuation

```python
from nltk.corpus import stopwords
import string
import re
```

### Stopwords in Japanese

```python
with open('japanese_stopwords.txt', 'r', encoding='utf-8') as file:
    stop_words_japanese = set(file.read().splitlines())
```

### Stopwords in Chinese

```python
import jieba

# Load Chinese stopwords from a file
with open('chinese_stopwords.txt', 'r', encoding='utf-8') as file:
    stop_words_chinese = set(file.read().splitlines())
```

### Stopwords in English

```python
stop_words_english = set(stopwords.words('english'))
```

### Apply stopwords and punctuation function

```python
# nltk.download('stopwords')

stop_words = stop_words_english.union(stop_words_chinese, stop_words_japanese)
punctuation = set(string.punctuation)
extra_punctuation = {'\ufeff', '‘', '’', '“', '”', '—', '–', '…'}
punctuation.update(extra_punctuation)

# Filter out stopwords, punctuation, and hidden characters
filtered_tokens = [
    word.strip('-') for word in tokens
    if word.strip('-').lower() not in stop_words
    and word not in punctuation
    and not re.match(r'[0-9\W]', word)  # Remove tokens with numbers or special characters
]

print("Filtered Tokens:", filtered_tokens)
```

## Stemming Words

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()

# Stem the filtered tokens
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print("Stemmed Tokens:", stemmed_tokens)
```

## Using BERT

```python
from transformers import TFBertTokenizer, TFBertModel
import tensorflow as tf
```

```python
import tensorflow as tf
import tensorboard

print("TensorFlow version:", tf.__version__)
print("TensorBoard version:", tensorboard.__version__)
```

```python
# Load pre-trained BERT model and tokenizer
tokenizer = TFBertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained('bert-base-uncased')

# Example diary text
text = " ".join(stemmed_tokens)

# Tokenize input
inputs = tokenizer(text, return_tensors='tf', truncation=True, padding=True)

# Get contextualized embeddings
outputs = model(**inputs)

# Extract embeddings for each token
embeddings = outputs.last_hidden_state
print("Embeddings shape:", embeddings.shape)
```
